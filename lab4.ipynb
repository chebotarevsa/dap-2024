6. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('titanic.csv', index_col='PassengerId')
young_group = df[df['Age'] < 30]
old_group = df[df['Age'] > 60]
young_survival_rate = young_group['Survived'].mean()
old_survival_rate = old_group['Survived'].mean()
print(f"Моложе 30: {young_survival_rate:.2f}")
print(f"Старше 60: {old_survival_rate:.2f}")
```

8.  Найдите самое популярное имя среди пассажиров Титаника мужского пола?

```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('titanic.csv', index_col='PassengerId')
male_passengers = df[df['Sex'] == 'male'].copy()
male_passengers.loc[:, 'First Name'] = male_passengers['Name'].str.extract(r'\.\s*([^ ]+)')[0]
most_common_name = male_passengers['First Name'].mode()[0]
name_count = male_passengers['First Name'].value_counts().loc[most_common_name]
print(f"Популярное имя '{most_common_name}' количество - {name_count}.")
```

10. Сравните возраст у спасенных и у погибших пасажиров. Средний возраст погибших выше, верно?

```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('titanic.csv')
df = df.dropna(subset=['Age'])
average_age_survived = df[df['Survived'] == 1]['Age'].mean()
average_age_died = df[df['Survived'] == 0]['Age'].mean()
print(f"Средний возраст спасенных пассажиров: {average_age_survived:.2f} лет")
print(f"Средний возраст погибших пассажиров: {average_age_died:.2f} лет")
if average_age_survived > average_age_died:
    print("Средний возраст спасенных выше среднего возраста погибших.")
else:
    print("Средний возраст погибших выше или равен среднему возрасту спасенных.")
```

### Дополнительные вопросы для защиты

1. Что такое медиана? Рассчитать медиану для предложенного ряда чисел.

Медиана — это центральное значение в упорядоченном ряду чисел. Если количество элементов нечётное, медианой будет средний элемент. Если количество элементов чётное, медиана — это среднее арифметическое двух центральных элементов.

2. Что такое дисперсия? Рассчитать дисперсию для предложенного ряда чисел.

Дисперсия — это мера разброса значений относительно их среднего (математического ожидания). Она показывает, насколько сильно элементы ряда отклоняются от среднего значения.

3. Что такое среднеквадратичное отклонение? Рассчитать среднеквадратичное отклонение для предложенного ряда чисел.

Среднеквадратичное отклонение — это квадратный корень из дисперсии, показывающий, как далеко элементы находятся от среднего значения.

4. Что такое Series и чем он отличается от DataFrame

Series — это одномерный массив с метками (индексами). Он может содержать данные любого типа, похож на столбец в таблице.

DataFrame — это двумерная структура данных (таблица), состоящая из нескольких объектов Series, каждый из которых представляет столбец. DataFrame может содержать данные разного типа в разных столбцах.

5. Можно ли изменить индекс в Series после создания объекта?

Да, можно. Индекс в объекте Series можно изменить с помощью атрибута .index или метода .set_index().

6. Можно ли выполнить операцию сложения над двумя объектами Series, если они имеют:
- одинаковые типы данных, одинаковые размеры и одинаковые индексы?
- одинаковые типы данных, одинаковые размеры, но разные индексы?,  
- одинаковые типы данных, разные размеры и одинаковые индексы?
- разные типы данных, одинаковые размеры и одинаковые индексы?
- разные типы данных, одинаковые размеры и разные индексы?
- разные типы данных, разные размеры и разные индексы?  
Ответ:
- Одинаковые типы данных, одинаковые размеры и одинаковые индексы — сложение возможно, результат будет также Series с теми же индексами.
- Одинаковые типы данных, одинаковые размеры, но разные индексы — сложение возможно, но объединение будет по индексам; отсутствующие значения будут заменены на NaN.
- Одинаковые типы данных, разные размеры и одинаковые индексы — сложение возможно, если длины совпадают, иначе будут NaN для несовпадающих индексов.
- Разные типы данных, одинаковые размеры и одинаковые индексы — сложение возможно, результат будет зависеть от типов данных.
- Разные типы данных, одинаковые размеры и разные индексы — сложение возможно, объединение по индексам, NaN для несовпадающих.
- Разные типы данных, разные размеры и разные индексы — сложение возможно, объединение по индексам, NaN для несовпадающих.
