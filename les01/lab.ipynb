7. Типы данных, рассказать подробно про один из типов данных (на выбор преподавателя).

```
# 1. Целые числа (int)
x = 10
print(f"x = {x}, тип данных: {type(x)}")  # <class 'int'>

# 2. Числа с плавающей точкой (float)
y = 10.5
print(f"y = {y}, тип данных: {type(y)}")  # <class 'float'>

# 3. Строки (str)
s = "Привет, мир!"
print(f"s = '{s}', тип данных: {type(s)}")  # <class 'str'>

# 4. Булевы значения (bool)
is_active = True
print(f"is_active = {is_active}, тип данных: {type(is_active)}")  # <class 'bool'>

# 5. Списки (list)
my_list = [1, 2, 3, 4, 5]
print(f"my_list = {my_list}, тип данных: {type(my_list)}")  # <class 'list'>

# 6. Кортежи (tuple)
my_tuple = (1, 2, 3, 4, 5)
print(f"my_tuple = {my_tuple}, тип данных: {type(my_tuple)}")  # <class 'tuple'>

# 7. Множества (set)
my_set = {1, 2, 3, 4, 5}
print(f"my_set = {my_set}, тип данных: {type(my_set)}")  # <class 'set'>

# 8. Словари (dict)
my_dict = {"name": "Alice", "age": 25}
print(f"my_dict = {my_dict}, тип данных: {type(my_dict)}")  # <class 'dict'>

# 9. NoneType
z = None
print(f"z = {z}, тип данных: {type(z)}")  # <class 'NoneType'>

```

8. Что такое Jupyter Notebook, для чего используется?

Jupyter Notebook — это интерактивная среда для создания и выполнения кода, которая поддерживает множество языков программирования (чаще всего Python) и используется для разработки, документирования и обмена проектами, связанными с данными.

9. Почему python популярен для анализа данных?

Простой и читаемый синтаксис, большое количество библиотек непосредственно для анализа данных, гибкость, кроссплатформенность, возможность интеграции с Jupyter Notebook и др.

10. Обязательно ли использовать Jupyter Notebook для анализа данных?

Использование Jupyter Notebook для анализа данных не является обязательным, хотя он очень популярен и удобен в этой сфере. Есть множество других инструментов и сред разработки, которые можно использовать для анализа данных, в зависимости от целей, предпочтений и рабочего процесса.
