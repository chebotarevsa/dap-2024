## Задача №1

```python
# Создать массив 10x10 со случайными значениями, найти минимум и максимум (низкая)
import numpy as np

# Генерирует матрицу 10x10 с элементами в диапазоне от 1 до 10 включительно
x = np.random.randint(1, 11, size=[10, 10])
print(F"MIN: {np.min(x)}, MAX: {np.max(x)}")
print("MATRIX:\n", x)
```

## Задача №2

```python
# Создать 8x8 матрицу и заполнить её единицами и нулями в шахматном порядке
import numpy as np

x = np.zeros((8, 8))
x[::2, ::2] = 1
x[1::2, 1::2] = 1
print(x)
```

## Задача №3

```python
# Дан массив из 11 элементов, поменять знак у элементов, значения которых между 3 и 8
import numpy as np

z = np.arange(11)
z[(3 < z) & (z < 8)] *= -1
print(z)
```

## Задача №4

```python
# Найти среднее значение строки матрицы и вычесть его из каждого элемента строки
import numpy as np

z = np.random.rand(5, 10)
print("Before:\n", z)
print("Mean for rows:\n", z.mean(1, keepdims=True))
z -= z.mean(1, keepdims=True)
print("After:\n", z)
```

## Задача №5

```python
# Поменяйте 1 и 2 строки матрицы
import numpy as np

z = np.arange(25).reshape(5, 5)
print("Before:\n", z)
z[[0, 1]] = z[[1, 0]]
# z[:2] = z[1::-1]
# z[0], z[1] = z[1].copy(), z[0].copy()
print("After:\n", z)
```

## Задача №6

```python
# Дан массив 10x2 представляющий собой точки в декартовой системе координат, преобразовать в полярную координату
import numpy as np

z = np.random.random((10, 2))
X = z[:, 0].copy()
Y = z[:, 1].copy()
print("Before:\n", z)
z[:, 0] = np.sqrt(pow(X, 2) + pow(Y, 2))
z[:, 1] = np.arctan2(Y, X)
print("After:\n", z)
```

## Задача №7

```python
# Заменить максимальный элемент массива на ноль
import numpy as np

z = np.random.random(10)
print(z)
print("Max:", z.max())
# z[z.argmax()] = 0 # Заменяет первое вхождение максимального элемента
z[z == z.max()] = 0  # Заменяем все максимальные элементы (если максимальный элемент встретился 2 раза)
print(z)
```

## Задача №8

```python
# Дан массив, добавить 1 к каждому элементу с индексом, заданным в другом массиве. Добавление производиться только 1 раз
import numpy as np

z = np.ones(10)
i = np.random.randint(0, len(z), 20)
print(z, i)
z[i] += 1
print(z)
```

## Задача №9

```python
# Дан вектор [1, 2, 3, 4, 5], построить новый вектор с тремя нулями между каждым значением
import numpy as np

z = np.array([1, 2, 3, 4, 5])
x = np.zeros((len(z) + (len(z) - 1) * 3), np.int64)
x[::4] = z
print(x)
```

## Задача №10

```python
# Найти ближайшее к заданному значению число в заданном массиве
import numpy as np

z = np.arange(100)  # Массив
v = np.random.randint(0, 100) # Число
print(z, v)
print(np.argwhere(z == v))
```
