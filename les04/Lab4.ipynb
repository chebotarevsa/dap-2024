```python
# Импорт необходимых библиотек
import numpy as np
import pandas as pd

# Загрузка данных из CSV файла и установка PassengerId в качестве индекса
df = pd.read_csv('titanic.csv', index_col='PassengerId')

# Просмотр первых 6 строк
print(df.head(6))

# Описание данных
print(df.describe())

# Сколько мужчин и женщин находилось на борту
sex_distribution = df['Sex'].value_counts()
print("Мужчин:", sex_distribution['male'], "Женщин:", sex_distribution['female'])

# Распределение Pclass и по полу, сколько мужчин 2-го класса
pclass_distribution = df['Pclass'].value_counts()
pclass_by_sex = df.groupby('Sex')['Pclass'].value_counts()

print("Распределение Pclass:")
print(pclass_distribution)
print("\nРаспределение Pclass для мужчин/женщин:")
print(pclass_by_sex)

# Сколько было мужчин 2-го класса
print("\nМужчин 2-го класса:", pclass_by_sex['male'][2])

# Медиана и среднеквадратичное отклонение платежей (Fare), округлено до 2 знаков
fare_median = round(df['Fare'].median(), 2)
fare_std = round(df['Fare'].std(), 2)

print(f"Медиана платежей: {fare_median}, Стандартное отклонение платежей: {fare_std}")

# Правда ли, что люди моложе 30 лет выживали чаще, чем старше 60?
younger_than_30 = df[df['Age'] < 30]
older_than_60 = df[df['Age'] > 60]

survival_rate_under_30 = round(younger_than_30['Survived'].mean() * 100, 2)
survival_rate_over_60 = round(older_than_60['Survived'].mean() * 100, 2)

print(f"Доля выживших младше 30 лет: {survival_rate_under_30}%, старше 60 лет: {survival_rate_over_60}%")

# Правда ли, что женщины выживали чаще мужчин? Доли выживших в обеих группах
survival_rate_men = round(df[df['Sex'] == 'male']['Survived'].mean() * 100, 2)
survival_rate_women = round(df[df['Sex'] == 'female']['Survived'].mean() * 100, 2)

print(f"Доля выживших среди мужчин: {survival_rate_men}%, среди женщин: {survival_rate_women}%")

# Самое популярное мужское имя
def get_first_name(fullname):
    name = fullname.split(',')[1].strip().split('.')
    if len(name) > 1:
        return name[1].strip().split(' ')[0]
    return ''

male_names = df[df['Sex'] == 'male']['Name'].apply(get_first_name)
most_common_name = male_names.value_counts().idxmax()

print(f"Самое популярное мужское имя: {most_common_name}")

# Средний возраст мужчин / женщин в зависимости от класса обслуживания
avg_age_by_class_and_sex = df.groupby(['Pclass', 'Sex'])['Age'].mean()

print("Средний возраст мужчин/женщин по классам обслуживания:")
print(avg_age_by_class_and_sex)

# В среднем мужчины 1-го класса старше 40 лет?
print("Мужчины 1-го класса старше 40 лет:", avg_age_by_class_and_sex[1]['male'] > 40)

# В среднем женщины 1-го класса старше 40 лет?
print("Женщины 1-го класса старше 40 лет:", avg_age_by_class_and_sex[1]['female'] > 40)

# Мужчины всех классов в среднем старше женщин того же класса?
for cls in [1, 2, 3]:
    print(f"Мужчины старше женщин в классе {cls}: {avg_age_by_class_and_sex[cls]['male'] > avg_age_by_class_and_sex[cls]['female']}")

# Люди в 1 классе старше, чем во 2-м и 3-м классах?
avg_age_by_class = df.groupby('Pclass')['Age'].mean()
print("Люди 1-го класса старше, чем во 2-ом и 3-ем?:", avg_age_by_class[1] > avg_age_by_class[2] > avg_age_by_class[3])

# Сравнение возраста у спасенных и погибших пассажиров
survivors_age = df[df['Survived'] == 1]['Age'].mean()
victims_age = df[df['Survived'] == 0]['Age'].mean()

print(f"Средний возраст спасенных: {round(survivors_age, 2)}, Средний возраст погибших: {round(victims_age, 2)}")
print("Средний возраст погибших выше?", victims_age > survivors_age)
```
