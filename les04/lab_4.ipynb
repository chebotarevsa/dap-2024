4. Выведите распределение переменной Pclass (социально-экономический статус) и это же распределение, только для мужчин / женщин по отдельности. Сколько было мужчин 2-го класса

6. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

```python
# Ваш код здесьimport pandas as pd
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('titanic.csv', index_col='PassengerId')
younger_than_30 = df[df['Age'] < 30]
older_than_60 = df[df['Age'] > 60]
survived_younger = younger_than_30['Survived'].sum()
total_younger = younger_than_30['Survived'].count()
survived_older = older_than_60['Survived'].sum()
total_older = older_than_60['Survived'].count()
survival_rate_younger = survived_younger / total_younger if total_younger > 0 else 0
survival_rate_older = survived_older / total_older if total_older > 0 else 0
print(f"Доля выживших среди людей младше 30 лет: {survival_rate_younger:.2f}")
print(f"Доля выживших среди людей старше 60 лет: {survival_rate_older:.2f}")
if survival_rate_younger > survival_rate_older:
    print("Люди младше 30 лет выживали чаще, чем люди старше 60 лет.")
else:
    print("Люди старше 60 лет выживали не чаще, чем люди младше 30 лет.")
```

7. Правда ли, что женщины выживали чаще мужчин? Каковы доли выживших в обеих группах?

```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('titanic.csv', index_col='PassengerId')
men = df[df['Sex'] == 'male']
women = df[df['Sex'] == 'female']
survived_men = men['Survived'].sum()
total_men = men['Survived'].count()
survived_women = women['Survived'].sum()
total_women = women['Survived'].count()
survival_rate_men = survived_men / total_men if total_men > 0 else 0
survival_rate_women = survived_women / total_women if total_women > 0 else 0
print(f"Доля выживших среди мужчин: {survival_rate_men:.2f}")
print(f"Доля выживших среди женщин: {survival_rate_women:.2f}")
if survival_rate_women > survival_rate_men:
    print("Женщины выживали чаще, чем мужчины.")
else:
    print("Мужчины выживали не реже, чем женщины.")
```

8.  Найдите самое популярное имя среди пассажиров Титаника мужского пола?

```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('titanic.csv', index_col='PassengerId')
male_passengers = df[df['Sex'] == 'male'].copy()
male_passengers.loc[:, 'First Name'] = male_passengers['Name'].str.extract(r'\.\s*([^ ]+)')[0]
most_common_name = male_passengers['First Name'].mode()[0]
name_count = male_passengers['First Name'].value_counts().loc[most_common_name]
print(f"Популярное имя '{most_common_name}' {name_count} совпадений.")
```

10. Сравните возраст у спасенных и у погибших пасажиров. Средний возраст погибших выше, верно?

```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('titanic.csv')
df = df.dropna(subset=['Age'])
average_age_survived = df[df['Survived'] == 1]['Age'].mean()
average_age_died = df[df['Survived'] == 0]['Age'].mean()
print(f"Средний возраст спасенных пассажиров: {average_age_survived:.2f} лет")
print(f"Средний возраст погибших пассажиров: {average_age_died:.2f} лет")
if average_age_survived > average_age_died:
    print("Средний возраст спасенных выше среднего возраста погибших.")
else:
    print("Средний возраст погибших выше или равен среднему возрасту спасенных.")
```
