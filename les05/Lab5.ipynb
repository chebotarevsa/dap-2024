###Задача 1: Построение двух линейных графиков

```python
import numpy as np
import matplotlib.pyplot as plt

# Диапазоны для x
x1 = np.linspace(-10, 0, 100)
x2 = np.linspace(0, 7, 100)

# Функции y
y1 = np.sin(x1)
y2 = np.sin(x2)

# Построение графиков
plt.plot(x1, y1, 'r-.', label="sin(x) для x [-10,0]")  # Красный цвет, штрихпунктирная линия
plt.plot(x2, y2, color=(1.0, 0.2, 0.3), label="sin(x) для x [0,7]")  # Сплошная линия, цвет RGB

# Установка лимитов
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Подписи и легенда
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Графики функции sin(x)')
plt.legend()

plt.show()
```

###Задача 2: Диаграмма рассеяния для набора данных Titanic

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv('titanic.csv')

# Удаление строк с пропусками в столбцах Age и Fare
df_clean = df.dropna(subset=['Age', 'Fare'])

# Разделение данных по выжившим и не выжившим
survived = df_clean[df_clean['Survived'] == 1]
not_survived = df_clean[df_clean['Survived'] == 0]

# Построение графика
plt.scatter(survived['Age'], survived['Fare'], color='green', label='Выжившие', marker='o')
plt.scatter(not_survived['Age'], not_survived['Fare'], color='red', label='Не выжившие', marker='x')

# Подписи
plt.xlabel('Возраст')
plt.ylabel('Стоимость билета')
plt.title('Диаграмма рассеяния: Возраст vs Стоимость билета')
plt.legend()

plt.show()
```

###Задача 3: Диаграммы рассеяния для набора Iris

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
import pandas as pd

# Загрузка данных Iris
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Создание парных диаграмм рассеяния
sns.pairplot(iris_df, hue='species')

# Добавление заголовка
plt.suptitle('Диаграммы рассеяния для набора Iris', y=1.02)

plt.show()
```

###Задача 4: Гистограммы для данных Titanic

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных Titanic
df = pd.read_csv('titanic.csv')

# Удаление строк с пропущенными значениями для возраста и стоимости билета
df_clean_age = df.dropna(subset=['Age'])
df_clean_fare = df.dropna(subset=['Fare'])

# Построение гистограммы по возрасту
plt.subplot(1, 2, 1)
plt.hist(df_clean_age['Age'], bins=range(0, 80, 5), edgecolor='black')
plt.title('Распределение пассажиров по возрасту')
plt.xlabel('Возраст')
plt.ylabel('Количество пассажиров')

# Гистограмма по стоимости билетов (только для женщин)
plt.subplot(1, 2, 2)
plt.hist(df_clean_fare[df_clean_fare['Sex'] == 'female']['Fare'], bins=20, color='purple', edgecolor='black')
plt.title('Распределение женщин по стоимости билетов')
plt.xlabel('Стоимость билета')
plt.ylabel('Количество пассажиров')

plt.tight_layout()
plt.show()
```

###Задача 5: Столбчатая диаграмма среднего Fare по классам

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv('titanic.csv')

# Средняя стоимость проезда по классам
mean_fare = df.groupby('Pclass')['Fare'].mean()

# Построение столбчатой диаграммы
mean_fare.plot(kind='bar', color='blue', edgecolor='black')

# Подписи
plt.xlabel('Класс каюты')
plt.ylabel('Средняя стоимость проезда')
plt.title('Средняя стоимость проезда для каждого класса')

plt.show()
```

###Задача 6: Столбчатая диаграмма с накоплением по полу

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv('titanic.csv')

# Подсчет пассажиров по классам и полу
pclass_sex = df.groupby(['Pclass', 'Sex']).size().unstack()

# Построение диаграммы с накоплением
pclass_sex.plot(kind='bar', stacked=True)

# Подписи
plt.xlabel('Класс каюты')
plt.ylabel('Количество пассажиров')
plt.title('Количество пассажиров по классам с разбивкой по полу')

plt.legend(title='Пол')

plt.show()
```

###Задача 7: Круговая диаграмма для набора Iris

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Загрузка данных Iris
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Подсчет количества каждого вида
species_counts = iris_df['species'].value_counts()

# Построение круговой диаграммы
plt.pie(species_counts, labels=iris.target_names, autopct='%1.1f%%', explode=[0.1, 0, 0], colors=['gold', 'lightblue', 'lightgreen'])

# Подписи
plt.title('Распределение видов ирисов')

plt.show()
```

###Задача 8: Гистограмма возраста пассажиров Titanic с использованием Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Загрузка данных Titanic
df = pd.read_csv('titanic.csv')

# Удаление строк с пропущенными значениями возраста
df_clean = df.dropna(subset=['Age'])

# Построение гистограммы
sns.histplot(df_clean['Age'], bins=20, kde=True)

# Подписи
plt.xlabel('Возраст')
plt.ylabel('Количество пассажиров')
plt.title('Распределение возраста пассажиров Титаника')

plt.show()
```
